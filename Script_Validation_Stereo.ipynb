#  Comparaison of the data from STERO PIV and ADV

The measurment by stereoscopic PIV gives the three componante of the velocity, It must by confirmed that those velocity are well measured from the data of the ADV wich take locally the velocity on (x,y,z) at 3 cm of the bottom


```python
# Import all the library: 
 
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from netCDF4 import Dataset
import cmocean
```

```python
# Importation of the function that manage the extraction of all the data from the `load_data.py' 
import load_data as data      # the py script `load_data.py' must be in the same folder than this notebook
```

```python
# Accéder aux variables spécifiques
u_exp13_adv = data.get_variable('EXP13', 'ADV', 'u')
#u_exp13_lab = data.get_variable('EXP13', 'lab', '')

u_exp13_adv
#print(u_exp13_lab)
```

```python
#Initialisation des variables temporelles et spatiales

time_exp13_stereo = data.get_variable('EXP13', 'STEREO', 'Time') 
time_exp13_adv    = data.get_variable('EXP13', 'ADV', 'time')+time_exp13_stereo[0]    # The beginning of the experiment was not set in the ADV file 

time_exp15_stereo = data.get_variable('EXP15', 'STEREO', 'Time') 
time_exp15_adv    = data.get_variable('EXP15', 'ADV', 'time')+time_exp15_stereo[0]    # The beginning of the experiment was not set in the ADV file 
```

```python


# Accéder aux variables spécifiques
u_exp13_adv    = data.get_variable('EXP13', 'ADV', 'v')    # Attention u et v sont intrevertis dans le fichier ADV
u_exp13_stereo = data.get_variable('EXP13', 'STEREO', 'U')
v_exp13_adv    = data.get_variable('EXP13', 'ADV', 'u')    # Attention u et v sont intrevertis dans le fichier ADV
v_exp13_stereo = data.get_variable('EXP13', 'STEREO', 'V')


u_exp15_adv    = data.get_variable('EXP15', 'ADV', 'v')    # Attention u et v sont intrevertis dans le fichier ADV
u_exp15_stereo = data.get_variable('EXP15', 'STEREO', 'U')
v_exp15_adv    = data.get_variable('EXP15', 'ADV', 'u')    # Attention u et v sont intrevertis dans le fichier ADV
v_exp15_stereo = data.get_variable('EXP15', 'STEREO', 'V')


```

```python
id_3cm = np.where((data.get_variable('EXP13', 'STEREO', 'coord_y'))==3.)
u_exp13_stereo[id_3cm].data

plt.figure(figsize=(15,8))
plt.plot(time_exp13_stereo,-u_exp13_stereo[id_3cm][0],label= 'Data STEREO')
plt.plot(time_exp13_adv,u_exp13_adv*100,label= 'Data ADV')
plt.grid()
#plt.yscale('log')
plt.xlabel('Time [s] ', fontsize=16)
plt.ylabel(r'U [cm.s$^{-1}$] ', fontsize=16)
plt.tick_params(axis='both', labelsize=14)
plt.legend(fontsize=14)
plt.show()

plt.figure(figsize=(15,8))
plt.plot(time_exp13_stereo,v_exp13_stereo[id_3cm][0],label= 'Data STEREO')
plt.plot(time_exp13_adv,-v_exp13_adv*100,label= 'Data ADV')
plt.grid()
#plt.yscale('log')
plt.xlabel('Time [s] ', fontsize=16)
plt.ylabel(r'V [cm.s$^{-1}$] ', fontsize=16)
plt.tick_params(axis='both', labelsize=14)
plt.legend(fontsize=14)
plt.show()
```
